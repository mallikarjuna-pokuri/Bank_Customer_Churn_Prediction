```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
churn = pd.read_csv('Churn_Modelling.csv')
churn.head()
```

```python
churn.describe()
```

```python
churn.Gender = churn.Gender.apply(lambda x:1 if x=='Male' else 0)
```

```python
churn = churn.drop(['CustomerId','Surname'],axis = 1)
```

```python
churn2 = pd.get_dummies(churn.Geography)
```

```python
churn = pd.concat([churn,churn2],axis = 1)
churn = churn.drop(['Geography','RowNumber'],axis = 1)
```

```python
churn.head()
```

```python
churn.Age = churn.Age.apply(lambda x:x/100)
churn.CreditScore = churn.CreditScore.apply(lambda x:x/850)
```

```python
churn.head()
churn['Exited'].value_counts()
```

```python

Y = churn.Exited
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(churn.drop('Exited',axis = 1),Y,stratify=Y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=2)
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```
